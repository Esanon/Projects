```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

restaurants = pd.read_csv('Datafiniti_Vegetarian_and_Vegan_Restaurants.csv')
```

```python
print(len(restaurants))
#10000 entries
print(restaurants.head(5))
```

```python
print(restaurants.columns)
```

```python
restaurants_info_needed = restaurants[["name","address" , "city","province","cuisines"]]
print(restaurants_info_needed.head())
```

```python
duplicates = restaurants_info_needed.duplicated()
print(duplicates.value_counts())
```

```python
print(restaurants_info_needed[1:100])
#a lot of duplicated restaurants. Will remove. 
```

```python
restaurants_info_needed = restaurants_info_needed.drop_duplicates()
print(len(restaurants_info_needed))
print(restaurants_info_needed.head())
```

```python
#How many restaurants are there in each metro area?
list_of_cities = restaurants_info_needed.city.unique()
print(list_of_cities)
```

```python
city_count = restaurants_info_needed.city.value_counts()
print(city_count)
```

```python
#How many restaurants in each metro area offers vegetarian options?
vegetarian_restaurants = restaurants_info_needed[restaurants_info_needed.cuisines == 'Vegetarian']
print(vegetarian_restaurants)
#doesn't show all restaurants.Need a different code. 
#only shows up as vegetarian if vegetarian is the only thing listed in cuisine. 
```

```python
vegetarian_restaurants_2 = restaurants_info_needed[restaurants_info_needed.cuisines.str.contains('Vegetarian')]
print(vegetarian_restaurants_2.head())
```

```python
Vegetarian_by_city = vegetarian_restaurants_2.city.value_counts()
print(Vegetarian_by_city)
#Result makes sense as all vegan restaurants are vegetarian so essentially all restaurants on list should be vegetarian.
```

```python
#How many restaurants in each metro area offers vegan options?
vegan_restaurants = restaurants_info_needed[restaurants_info_needed.cuisines.str.contains('Vegan')]
print(vegan_restaurants.head())
```

```python
vegan_by_city = vegan_restaurants.city.value_counts()
print(vegan_by_city)
```
