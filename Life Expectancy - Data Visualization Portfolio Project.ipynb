```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('all_data.csv')
print(df.head())
```

```python
df = df.rename({'Life expectancy at birth (years)':'Life'}, axis = 'columns')
```

```python
#Possible questions to answer:
#Has life expectancy increased over time in the six nations? check 
#Has GDP increased over time in the six nations?  check
#Is there a correlation between GDP and life expectancy of a country? check
#What is the average life expectancy in these nations? check
```

```python
print(df['Country'].unique())
```

```python
chile = df[df['Country'] == 'Chile']
china = df[df['Country'] == 'China']
germany = df[df['Country'] == 'Germany']
mexico = df[df['Country'] == 'Mexico']
usa = df[df['Country'] == 'United States of America']
zimbabwe = df[df['Country'] == 'Zimbabwe']
print(zimbabwe) #double check
```

```python
#Average life expectancies and GDPs 
print('The average life expectancy for Chile is ' + str(np.mean(chile['Life'])))
print('The average life expectancy for China is ' + str(np.mean(china['Life'])))
print('The average life expectancy for Germany is ' + str(np.mean(germany['Life'])))
print('The average life expectancy for Mexico is ' + str(np.mean(mexico['Life'])))
print('The average life expectancy for the United States of America is ' + str(np.mean(usa['Life'])))
print('The average life expectancy for Zimbabwe is ' + str(np.mean(zimbabwe['Life'])))
```

```python
#Shorter option. Year not needed. 
average_life = df.groupby(df['Country']).mean().reset_index().drop(columns = 'Year')
print(average_life)
```

```python
#graph the average life expectancies
plt.figure(figsize= (10,8))
sns.barplot(data = average_life, x = 'Country', y='Life')
plt.title('Average Life Expectancy at Birth by Country')
```

```python
#graph the average gdp's 
plt.figure(figsize= (10,8))
sns.barplot(data = average_life, x = 'Country', y='GDP')
plt.title('Average GDP by Country')
plt.ylabel('In Billions USD')
```

```python
#Life Expectancy increase? - line graphs 
plt.figure(figsize= (10,8))
sns.set_palette('bright')
sns.lineplot(x = df.Year, y= df.Life, hue= df.Country)
plt.title('2000-2015 Life Expectancy')
```

```python
print("With the exception of Zimbabwe, the other countries had small linear growths. Given the limited capabilities of human life expectancy and the starting points in 2000, this makes sense that the only country with a substantial amount of growth was Zimbabwe. Zimbabwe had a much lower life expectancy in 2000 and the increase from 46 to 60 is much more attainable. As Zimbabwe's GDP grows and the country continues to develop, one can expect the life expetancy to continually increase.") 
```

```python
#individual country plots for better visualization 

plt.figure(figsize= (10,8))
plt.subplot(2,3,1)
plt.plot(chile.Year, chile.Life, color='green')
plt.title('Chile Life Expectancy')

plt.subplot(2,3,2)
plt.plot(china.Year, china.Life, color='red')
plt.title('China Life Expectancy')

plt.subplot(2,3,3)
plt.plot(germany.Year, germany.Life, color='blue')
plt.title('Germany Life Expectancy')

plt.subplot(2,3,4)
plt.plot(mexico.Year, mexico.Life, color='yellow')
plt.title('Mexico Life Expectancy')

plt.subplot(2,3,5)
plt.plot(usa.Year, usa.Life, color='black')
plt.title('United States Life Expectancy')

plt.subplot(2,3,6)
plt.plot(zimbabwe.Year, zimbabwe.Life, color='purple')
plt.title('Zimbabwe Life Expectancy')

plt.subplots_adjust(top=0.95, hspace=0.75, wspace= .75)
plt.show()

```

```python
#same graphs to check for gdp increase
plt.figure(figsize= (10,8))
sns.set_palette('pastel')
sns.lineplot(x = df.Year, y= df.GDP, hue= df.Country)
plt.title('2000 - 2015 GDP by Country')
```

```python
#individual subplots for better visualization 
plt.figure(figsize= (10,8))
plt.subplot(2,3,1)
plt.plot(chile.Year, chile.GDP, color='green')
plt.title('Chile GDP')

plt.subplot(2,3,2)
plt.plot(china.Year, china.GDP, color='red')
plt.title('China GDP')

plt.subplot(2,3,3)
plt.plot(germany.Year, germany.GDP, color='blue')
plt.title('Germany GDP')

plt.subplot(2,3,4)
plt.plot(mexico.Year, mexico.GDP, color='yellow')
plt.title('Mexico GDP')

plt.subplot(2,3,5)
plt.plot(usa.Year, usa.GDP, color='black')
plt.title('United States GDP')

plt.subplot(2,3,6)
plt.plot(zimbabwe.Year, zimbabwe.GDP, color='purple')
plt.title('Zimbabwe GDP')

plt.subplots_adjust(top=0.95, hspace=0.75, wspace= .75)
plt.show()
```

```python
print('GDP has increased over time between 2000 and 2015 for every country. Three of the countries: Chile, Germany and Mexico,  however seem to have a decreasing GDP as of 2012. China has had the most linear growth visually as opposed to the Zimbabwe and the United States which both had slight dips.')
```

```python
#Scatterplots for correlation between life expectancy and GDP. 
plt.figure(figsize= (10,8))
sns.set_palette('Accent')
sns.scatterplot(x = df.Year, y= df.GDP, hue= df.Country)
plt.title('2000 - 2015 GDP by Country')
```

```python
sns.scatterplot(x=chile.Life, y=chile.GDP)
plt.title('Chile GDP-Life expectancy Correlation')
```

```python
#individual scatter plots 

plt.figure(figsize= (14,10))
plt.subplot(2,3,1)
sns.scatterplot(x=chile.Life, y=chile.GDP, marker='s')
plt.title('Chile Correlation')

plt.subplot(2,3,2)
sns.scatterplot(x=china.Life, y=china.GDP, marker='s')
plt.title('China Correlation')

plt.subplot(2,3,3)
sns.scatterplot(x=germany.Life, y= germany.GDP, marker='s')
plt.title('Germany Correlation')

plt.subplot(2,3,4)
sns.scatterplot(x=mexico.Life, y=mexico.GDP, marker='s')
plt.title('Mexico Correlation')

plt.subplot(2,3,5)
sns.scatterplot(x=usa.Life, y=usa.GDP, marker='s')
plt.title('US Correlation')

plt.subplot(2,3,6)
sns.scatterplot(x=zimbabwe.Life, y= zimbabwe.GDP, marker='s')
plt.title('Zimbabwe Correlation')

plt.subplots_adjust(top=.6, hspace=0.8, wspace= .8)
plt.show()
```

```python
print('As expected, there seems to be a positive correlation between the listed countries GDP and their life expectancy.')
```
